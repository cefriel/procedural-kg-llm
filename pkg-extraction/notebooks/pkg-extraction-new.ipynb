```python
# prompt 1 (P1)
from langchain.chains import LLMChain
from langchain.llms import OpenAI
from langchain.prompts import PromptTemplate
from langchain.chat_models import ChatOpenAI
from langchain.chains import SequentialChain

template ="""
You are an expert in information extraction, with a special background in procedures.

###
Consider the following procedure:
""
{example}
""

Q: generate sentences that include the verb/phrasal verb/idiomatic expression expressing the activity, the direct object on which the activity is performed (if any), the equipment that is needed for performing the activity (if any), except for the direct objects of the activity, details about time (i.e. duration or exact date or time) relevant for performing the activity (if any) 
A:
{example_extraction}
###

###
Consider the following process:
""
{procedure_for_llm}
""
Q: generate sentences that include the verb/phrasal verb/idiomatic expression expressing the activity, the direct object on which the activity is performed (if any), the equipment that is needed for performing the activity (if any), except for the direct objects of the activity, details about time (i.e. duration or exact date or time) relevant for performing the activity (if any) 
A:

"""

prompt = PromptTemplate(
    input_variables=["procedure_for_llm"],
    template = template                      
)

chain1 = LLMChain(
#    llm=ChatOpenAI(model="gpt-3.5-turbo-16k", temperature=0),
   llm=ChatOpenAI(model="gpt-4o", temperature=0),
    prompt=prompt,
    output_key="test"
)

overall_chain = SequentialChain(
    chains=[chain1],
    input_variables=["procedure_for_llm", "example", "example_extraction"],
    output_variables=["test"],
    verbose=True
)

output = overall_chain({"procedure_for_llm": procedure_for_llm, "example": keyboard_example, "example_extraction": keyboard_extraction})

#print(procedure_for_llm + "\n" + "\n")
print(output["test"])
```

```python
#- Detect direct objects that have been wrongly annotated together, even if they refer to different objects, and recommend to annotate them as separate direct objects.
#- Detect equipments that have been wrongly annotated together, even if they refer to different equipments, and recommend to annotate them as separate equipments.
#- Detect any missing entities (agent, activity, direct object, equipment) that have not been annotated, and recommend to include them in the annotation.
#- Detect direct objects that have been wrongly split as multiple objects, even if they refer to the same object (e.g. "the base of the microscope" is a unique object, do not split "base" and "microscope"), and recommend to put them as one direct objecct.
#- Detect direct objects that have been wrongly annotated together, even if they refer to different objects, and recommend to annotate them as separate direct objects.
#- Detect equipments that have been wrongly annotated together, even if they refer to different equipments, and recommend to annotate them as separate equipments.

from langchain.chains import LLMChain
from langchain.llms import OpenAI
from langchain.prompts import PromptTemplate
from langchain.chat_models import ChatOpenAI
from langchain.chains import SequentialChain

template ="""
You are an expert in information extraction, with a special background in procedures, capable of analysing entities extracted from a procedure and suggest changes that can improve the extraction.
You are not supposed to generate a new information extraction, just provide suggestions for changes when pertinent.
You always pay extra attention at the following:
- Detect direct objects of activities that have not been annotated also as equipment, even if they are material things needed for the activity, and recommend they are repeated also as equipment.
- Detect activities where the phrasal verb has been wrongly reported only as a verb without its respective adverb or preposition, and recommend that the adverb or preposition, if any, is added to the verb of the activity.
- Detect any missing entities (agent, activity, direct object, equipment) that have not been annotated, and recommend to include them in the annotation.
- Detect annotations that have been wrongly split even if they refer to the same thing (e.g. "base of microscope" is a unique object), and recommend to fix it.
- Detect annotations that have been wrongly merged even if they refer to different things, and recommend to fix it.
You don't always need to propose changes, if the extracted entities are good as-is just do not propose changes.

Analyse the extracted entities:
""
{test}
""
"""

prompt = PromptTemplate(
    input_variables=["test"],
    template = template                      
)

chain2 = LLMChain(
    llm=ChatOpenAI(model="gpt-3.5-turbo-16k", temperature=0),
#    llm=ChatOpenAI(model="gpt-4", temperature=0),
    prompt=prompt,
    output_key="corrections"
)

overall_chain = SequentialChain(
    chains=[chain1, chain2],
    input_variables=["procedure_for_llm", "example", "example_extraction"],
    output_variables=["test", "corrections"],
    verbose=True
)

output = overall_chain({"procedure_for_llm": procedure_for_llm, "example": keyboard_example, "example_extraction": keyboard_extraction})

##print(procedure_for_llm + "\n" + "\n")
print(output["test"] + "\n" + "\n")
print(output["corrections"] + "\n" + "\n")
```

```python
from langchain.chains import LLMChain
from langchain.llms import OpenAI
from langchain.prompts import PromptTemplate
from langchain.chat_models import ChatOpenAI
from langchain.chains import SequentialChain

template ="""
You are an expert in information extraction, with a special background in procedures, capable of modifying entities extracted from a procedure.
You apply some changes and then return the modified in the same format as the original one.
You apply exclusively the changes you're given and only the changes you're given. No more.

###
Consider the following extracted entities:
""
{test}
""
and apply the following changes:
""
{corrections}
""
"""

prompt = PromptTemplate(
    input_variables=["corrections"],
    template = template                      
)

chain3= LLMChain(
    llm=ChatOpenAI(model="gpt-3.5-turbo-16k", temperature=0),
#    llm=ChatOpenAI(model="gpt-4", temperature=0),
    prompt=prompt,
    output_key="corrected-entities"
)

overall_chain = SequentialChain(
    chains=[chain1, chain2, chain3],
    input_variables=["procedure_for_llm", "example", "example_extraction"],
    output_variables=["test", "corrections", "corrected-entities"],
    verbose=True
)

output = overall_chain({"procedure_for_llm": procedure_for_llm, "example": keyboard_example, "example_extraction": keyboard_extraction})

#print(procedure_for_llm + "\n" + "\n")
print(output["test"] + "\n" + "\n")
print(output["corrections"] + "\n" + "\n")
print(output["corrected-entities"] + "\n" + "\n")
```

```python
from langchain.chains import LLMChain
from langchain.llms import OpenAI
from langchain.prompts import PromptTemplate
from langchain.chat_models import ChatOpenAI
from langchain.chains import SequentialChain

template ="""
You are an expert in knowledge graph construction, with a special background in ontologies on procedural knowledge.

###
Consider this previous result:
""
{example1}
""

Q:  generate the RDF knowledge graph in Turtle format from it
A:
{example2}

###

###
Consider your previous result:
""
{test}
""
Q:  generate the RDF knowledge graph in Turtle format from it, based on the example
A:

"""

prompt = PromptTemplate(
    input_variables=["procedure_for_llm"],
    template = template                      
)

chain2 = LLMChain(
    llm=ChatOpenAI(model="gpt-3.5-turbo-16k", temperature=0),
#    llm=ChatOpenAI(model="gpt-4", temperature=0),
    prompt=prompt,
    output_key="test2"
)

overall_chain = SequentialChain(
    chains=[chain1, chain2],
    input_variables=["procedure_for_llm", "example1", "example2"],
    output_variables=["test", "test2"],
    verbose=True
)

output = overall_chain({"procedure_for_llm": procedure_for_llm, "example1": example1, "example2": example2})

#print(procedure_for_llm + "\n" + "\n")
print(output["test"] + "\n" + "\n")
print(output["test2"] + "\n" + "\n")
```
